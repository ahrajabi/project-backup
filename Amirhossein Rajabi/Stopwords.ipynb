```python
from bs4 import BeautifulSoup as bs
import requests
import redis
import time
import numpy as np
import lda
import lda.datasets
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
r = redis.StrictRedis()
swfa = pd.read_csv('../dataset/stopwords.csv' , header=None )
```

```python
start_time = time.time()

from hazm import *

clean_train = []
clean_train_original = []
cnt = 0
for i in r.keys(pattern='irna#821200**'):
    body = r.get(i)
    cnt += 1
    if cnt % 50 == 0:
        print(cnt)
    if body == 'None':
        continue
    body = body.decode('utf-8')
    norm = Normalizer()
    stemmer = Stemmer()
    lemmatizer = Lemmatizer()
    body = norm.normalize(body)
    body = word_tokenize(body)
    body = [w for w in body if not w in list(swfa[0])]
    clean_train_original.append(body)
    clean_train.append( " ".join( body ) )
print("--- %s seconds ---" % ( time.time() - start_time ))
```

```python
vectorizer = CountVectorizer(analyzer = "word",   \
                             tokenizer = None,    \
                             preprocessor = None, \
                             stop_words = None,   \
                             max_features = 500)
train_data_features_original = vectorizer.fit_transform(clean_train)
train_data_features = train_data_features_original.toarray()
vocab = vectorizer.get_feature_names()
dist = np.sum(train_data_features, axis=0)
listvocab = []
for i , j in enumerate(vocab):
    listvocab.append((list(dist)[i],j))
listvocab = sorted(listvocab,reverse=True)
for i in listvocab:
    print(i[1])
```
